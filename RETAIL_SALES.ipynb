# **QUERY**

```python
sales_df['Week'] = sales_df['Date'].dt.strftime('%U')
```

*   The strftime() is used to convert a datetime object into a string representation based on a specified format.

*   %U is a directive in strftime() that represents the week number of the year, with Sunday as the first day of the week (0-53).

*  %U considers the week starting from Sunday.

*  %W considers the week starting from Monday.





```python
sales_df['Week']
```

```python
weekly_sales_in_store = sales_df.groupby(['Store', 'Week'])['Weekly_Sales'].sum().reset_index()
```

```python
weekly_sales_in_store
```

```python
highest_sales_week = weekly_sales_in_store.loc[weekly_sales_in_store.groupby('Week')['Weekly_Sales'].idxmax()].reset_index()
```

```python
highest_sales_week
```

# **Which store performs high sales in week wise**

```python
#  to convert a datetime object into a string

sales_df['Week'] = sales_df['Date'].dt.strftime('%U')

# calculating the sum of weekly sales as per the query

weekly_sales_in_store = sales_df.groupby(['Store', 'Week'])['Weekly_Sales'].sum().reset_index()

# analysis the highest sales in the week

highest_sales_week = weekly_sales_in_store.loc[weekly_sales_in_store.groupby('Week')['Weekly_Sales'].idxmax()].reset_index()

highest_sales_week  # query satisfied
```
